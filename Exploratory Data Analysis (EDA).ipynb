# Exploratory Data Analysis (EDA)

```python
# Importing library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Loding data from datasets

```python
df = pd.read_csv('dataset.csv')
df
```

### Checking for null values in Dataset

```python
df.isnull().sum()
```

```python
# As we have two columns has null entries 
# Both columns doesn't have any importance for dataset so we can drop both columns

df.dropna(how = 'all' , axis = 1 , inplace = True)
df
```

```python
labels = ['Married', 'Single', 'Divorced']
explode = [0.05,0,0]

plt.pie(pd.value_counts(df['MaritalStatus']),labels = labels , autopct = '%.3f%%' , 
        explode = explode , startangle = 135 , counterclock = True)
plt.axis('equal')
```
