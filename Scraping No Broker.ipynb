## Scraping Data for Mumbai Flats

```python
res = requests.get('https://www.nobroker.in/property/sale/mumbai/multiple?searchParam=W3sibGF0IjoxOC45NjkwMjQ3LCJsb24iOjcyLjgyMDUyOTIsInBsYWNlSWQiOiJDaElKTjNHeG9XN081enNSNF9YTE83R09HZjQiLCJwbGFjZU5hbWUiOiJNdW1iYWkgQ2VudHJhbCIsInNob3dNYXAiOmZhbHNlfSx7ImxhdCI6MTkuMTUzODIzMSwibG9uIjo3Mi44NzUxNzg2LCJwbGFjZUlkIjoiQ2hJSkgyR3h0aF9JNXpzUjg2RGZhOXMtQzN3IiwicGxhY2VOYW1lIjoiTXVtYmFpIFN1YnVyYmFuIiwic2hvd01hcCI6ZmFsc2V9LHsibGF0IjoxOC45NzA2MDE0LCJsb24iOjcyLjgyMjAwNjM5OTk5OTk5LCJwbGFjZUlkIjoiQ2hJSnFmRVp4UlhPNXpzUkM0ZnYwZ0JRaXZnIiwicGxhY2VOYW1lIjoiTXVtYmFpIFNvdXRoIiwic2hvd01hcCI6ZmFsc2V9XQ==&radius=2.0&type=BHK2,BHK3&city=mumbai&locality[0]=Mumbai%20Central,&locality[1]=Mumbai%20Suburban,&locality[2]=Mumbai%20South')

res
```

```python
soup = BeautifulSoup(res.content , 'html.parser')
soup.prettify
```

```python
soup.title.text
```

```python
flats = soup.find_all('a', class_='overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full')
```

```python
flat_names = []
for i in flats:
    flat_names.append(i.text)

flat_names
```

```python
bhk = []
for i in flat_names:
    bhk.append(i[:5])
    
print(bhk , end= ' ')
```

```python
name = []
for names in flat_names:
    name.append(names[6:])

name
```

```python
flat_prices = soup.find_all('div', "font-semi-bold heading-6")
```

```python
prices = []

for i in flat_prices:
    prices.append(i.text)
    
prices
```

```python
rate = []
emi = []
area = []

for j in range(0, len(prices),3):
        rate.append(prices[j])
    
for k in range(1, len(prices), 3):
        emi.append(prices[k])

for l in range(2, len(prices) ,3):
        area.append(prices[l])

        
# len(area), len(rate), len(emi) , len(name) , len(bhk)
```

```python
df = pd.DataFrame({'Flat Type': bhk , 'Address': name , 'Price': rate , 'EMI': emi , 'Area': area})

df
```

```python
df.set_index(np.arange(1,26) , inplace = True)

df
```

#### Exporting Data to Excel File

```python
df.to_csv('Mumbai Flats.csv')
```
